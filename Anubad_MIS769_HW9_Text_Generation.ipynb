## Part 2: Parameter Exploration (4 points)

````python
# Temperature comparison
print("📊 TEMPERATURE COMPARISON")
print("=" * 60)

prompt = "Once upon a time"
temperatures = [0.3, 0.7, 1.2]

for temp in temperatures:
    output = generator(
        prompt,
        max_new_tokens=30,
        temperature=temp,
        do_sample=True,
        pad_token_id=50256
    )
    generated = output[0]['generated_text'][len(prompt):]
    print(f"\nTemperature {temp}:")
    print(f"  {generated[:80]}...")
````

````python
# Top-k and Top-p comparison
print("📊 TOP-K AND TOP-P COMPARISON")
print("=" * 60)

configs = [
    {"top_k": 10, "top_p": 1.0, "name": "top_k=10"},
    {"top_k": 50, "top_p": 1.0, "name": "top_k=50"},
    {"top_k": 0, "top_p": 0.9, "name": "top_p=0.9"},
    {"top_k": 0, "top_p": 0.5, "name": "top_p=0.5"},
]

for cfg in configs:
    output = generator(
        prompt,
        max_new_tokens=30,
        top_k=cfg["top_k"] if cfg["top_k"] > 0 else None,
        top_p=cfg["top_p"],
        do_sample=True,
        pad_token_id=50256
    )
    generated = output[0]['generated_text'][len(prompt):]
    print(f"\n{cfg['name']}:")
    print(f"  {generated[:80]}...")
````

### The Repetition Problem

GPT-2 is notorious for getting stuck in loops — repeating the same phrase or sentence over and over. This happens because the model assigns high probability to tokens it has already generated (a feedback loop).

**Repetition penalty** (`repetition_penalty`) solves this by reducing the probability of tokens that have already appeared. A value of `1.0` means no penalty (default), while `1.2`-`1.5` progressively discourages repetition.

```
repetition_penalty = 1.0  →  "The cat sat on the mat. The cat sat on the mat. The cat..."
repetition_penalty = 1.3  →  "The cat sat on the mat. It purred softly as sunlight streamed..."
```

This is especially important for **longer generations** (100+ tokens) where loops are more likely to appear. Try running the same prompt with and without the penalty to see the difference.

````python
# Repetition penalty comparison
print("📊 REPETITION PENALTY COMPARISON")
print("=" * 60)

prompt = "The company announced that"
penalties = [1.0, 1.3]

for penalty in penalties:
    output = generator(
        prompt,
        max_new_tokens=80,
        temperature=0.7,
        repetition_penalty=penalty,
        do_sample=True,
        pad_token_id=50256
    )
    generated = output[0]['generated_text'][len(prompt):]
    label = "No penalty" if penalty == 1.0 else f"Penalty {penalty}"
    print(f"\n{label} (repetition_penalty={penalty}):")
    print(f"  {generated[:120]}...")
````

## Part 3: Gradio Interface (5 points)

````python
import gradio as gr

def generate_text(prompt, temperature, max_length):
    """Generate text with specified parameters."""
    if not prompt:
        return "Please enter a prompt."

    result = generator(
        prompt,
        max_new_tokens=int(max_length),
        temperature=temperature,
        do_sample=True,
        pad_token_id=50256
    )
    return result[0]['generated_text']

# Create Gradio interface
demo = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(label="Prompt", placeholder="Enter your prompt here...", lines=2),
        gr.Slider(0.1, 1.5, value=0.7, step=0.1, label="Temperature"),
        gr.Slider(20, 200, value=50, step=10, label="Max New Tokens")
    ],
    outputs=gr.Textbox(label="Generated Text", lines=5),
    title="🚀 Text Generator",
    description="Generate creative text using GPT-2. Adjust temperature for creativity.",
    examples=[
        ["The future of technology is", 0.7, 50],
        ["Once upon a time in a magical forest", 0.9, 100],
        ["Dear customer, thank you for", 0.5, 50]
    ]
)

print("🎨 GRADIO INTERFACE CREATED")
print("=" * 60)
print("Run demo.launch() to start the interface")
````

````python
# Launch the demo (in Colab, this creates a public URL)
demo.launch(share=True)
````

## Part 4: HuggingFace Spaces Deployment (5 points)

````python
# Generate the app.py file for Spaces
app_code = '''
import gradio as gr
from transformers import pipeline

# Load model
generator = pipeline("text-generation", model="gpt2")

def generate_text(prompt, temperature, max_length):
    if not prompt:
        return "Please enter a prompt."

    result = generator(
        prompt,
        max_new_tokens=int(max_length),
        temperature=temperature,
        do_sample=True,
        pad_token_id=50256
    )
    return result[0]['generated_text']

demo = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(label="Prompt", placeholder="Enter your prompt..."),
        gr.Slider(0.1, 1.5, value=0.7, label="Temperature"),
        gr.Slider(20, 200, value=50, label="Max Tokens")
    ],
    outputs=gr.Textbox(label="Generated Text"),
    title="Text Generator",
    description="Generate text with GPT-2"
)

demo.launch()
'''

print("📄 APP.PY FOR HUGGINGFACE SPACES")
print("=" * 60)
print(app_code)

# Creating app.py file for HuggingFace Spaces
print("Creating app.py file for HuggingFace Spaces...")
with open("app.py", "w") as f:
    f.write(app_code)
    print("✅ app.py file created successfully!")
````

````python
# Generate requirements.txt
requirements = """transformers
gradio
torch
"""

print("📄 REQUIREMENTS.TXT")
print("=" * 60)
print(requirements)

# Creating requirements.txt file for HuggingFace Spaces
print("Creating requirements.txt file for HuggingFace Spaces...")
with open("requirements.txt", "w") as f:
    f.write(requirements)
    print("✅ requirements.txt file created successfully!")
````

**Q1:** How do temperature and top-p affect output creativity?

**A1:** Both control the randomness of the model, but for different outcomes:
* **Temperature** flattens the probability distribution. At high temperatures, even "unlikely" words get a dcent chance to be picked, making the output more diverse (at high values) or selective (at low values)
* **Top-*P*** acts as a dynamic safety net, ensuring that the model only chooses from a set of words that are actually plausible, cutting off the "tail" of irrelevant words (*tail* refers to the extreme ends of the probability distribution).
<bnr>Together, **temperature** and **top-*P*** balance creativity with coherency.
